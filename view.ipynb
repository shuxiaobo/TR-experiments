# 错误样例分析

```python
pred = list()
origin_dev = list()
with open(os.path.join(args.tmp_dir, 'AIC', args.valid_file)) as f:
    for line in f:
        origin_dev.append(json.loads(line))
with open('tf/weights/submissiondev.txt') as f:
    for line in f:
        s =  line.strip().split('\t')
        pred.append([int(s[0]), s[1]])

origin_dev[0]
pred[0]
```

```python
for i, d in enumerate(origin_dev):
    if d['query_id'] != pred[i][0]:
        raise Exception("dsad")
    elif d['answer'] != pred[i][1]:
        print(d)
```

```python
for i, d in enumerate(origin_dev):
    if d['query_id'] != pred[i][0]:
        raise Exception("dsad")
    elif d['answer'] != pred[i][1]:
        print(pred[i][1]+'**'*3 + str(len(d['passage']))+'**'*3+d['answer'])
```

```python
origin_dev_answers = [d['answer'] for d in origin_dev]
pred_answers = [d[1] for d in pred]
```
